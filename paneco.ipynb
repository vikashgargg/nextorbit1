```python
paneco = pd.read_csv('paneco1.csv')
cts = pd.read_csv('cts_brand.csv')
```

```python
paneco['Brand']=paneco['Brand']
paneco['Brand'].nunique()
```

```python
paneco['Brand']=paneco['Brand'].str.lower()
cts['cts_brand']=cts['cts_brand'].str.lower()
```

```python
l=cts['cts_brand'].tolist()
l
```

```python
def match(Brand):
    Brand=str(Brand).lower()
    for cts_brand in l:
        if Brand in cts_brand.lower():
            return cts_brand
    return '-'
    
```

```python
paneco['brand_match']=paneco['Brand'].apply(match)
```

```python
paneco['brand_match'].nunique()
```

```python
def nomatch(brand):
    brand=str(brand).lower()
    for cts_brand in l:
        if brand in cts_brand.lower():
            return brand
    return '-'
```

```python
paneco['brand_match2']=paneco['Brand'].apply(nomatch)
paneco['brand_match2'].unique()
```

```python
paneco['missing_brands']=paneco['Brand'][~paneco['Brand'].isin(paneco['brand_match2'])]
```

```python
paneco['missing_brands'].nunique()
```
