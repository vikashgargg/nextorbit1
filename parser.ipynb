```python
ll = []
for o in l:
    try:
        temp = {}
        temp['url']  = o['url']
        temp['api']  = o['api']
        temp['name'] = o[':items']['root'][':items']['responsivegrid'][':items']['productpage']['product']['productDisplayName']
        temp['discount_price'] = o[':items']['root'][':items']['responsivegrid'][':items']['productpage']['product']['variantOptions'][0]['offers'][0]['channelPrices'][0]['enDiscountedValue']
        temp['price'] = o[':items']['root'][':items']['responsivegrid'][':items']['productpage']['product']['variantOptions'][0]['offers'][0]['channelPrices'][0]['enOriginalValue']
        temp['brand'] = o[':items']['root'][':items']['responsivegrid'][':items']['productpage']['product']['brandCategories'][0]['name']
        temp['description'] = o[':items']['root'][':items']['responsivegrid'][':items']['productpage']['product']['description']
        temp['oos'] = o[':items']['root'][':items']['responsivegrid'][':items']['productpage']['product']['stockStatus'] #NONE
        temp['oos_2'] = o[':items']['root'][':items']['responsivegrid'][':items']['productpage']['product']['variantOptions']

        temp['name_in_english'] = o[':items']['root'][':items']['responsivegrid'][':items']['productpage']['product']['nameInEnglish'] 
        ll.append(temp)
    except Exception as e:
        pass
```
